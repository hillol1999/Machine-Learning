```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel('https://github.com/YBIFoundation/MarketBasketAnalysis/raw/main/DataSet/CoffeShopSales.xlsx')
df.head()
```

```python
df
```

```python
df_pivot = df.pivot_table(index='TransactionID', columns='Item', values='Quantity', aggfunc='sum').fillna(0)
```

```python
df_pivot
```

```python
df_pivot = df_pivot.astype(int)
df_pivot
```

```python
def encode (x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
encode(10)
```

```python
encode(-10)
```

```python
df_pivot = df_pivot.applymap(encode)
df_pivot
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
frequent_items = apriori(df_pivot, min_support=0.01, max_len=2, use_colnames=True)
```

```python
association_rules(frequent_items, metric="lift", min_threshold=1)
```

```python
df_dashboard = association_rules(frequent_items, metric="lift", min_threshold=1)
```

```python
df_dashboard.to_csv('dashboard')
```
