# Admission Chance

```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/f83fe655797d5d59f8c36d3df9fc09bd1fbfe4b4/Admission%20Chance.csv')
df
```

```python
df.columns
```

```python
y = df['Chance of Admit ']
x = df[['GRE Score', 'TOEFL Score', 'University Rating', ' SOP', 'LOR ', 'CGPA', 'Research']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2529)
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train_ss = ss.fit_transform(x_train)
x_test_ss = ss.transform(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x_train_mms = mms.fit_transform(x_train)
x_test_mms = mms.transform(x_test)
```

```python
x_train.describe()
```

```python
pd.DataFrame(x_train_ss).describe()
```

```python
pd.DataFrame(x_train_mms).describe()
```

```python
from sklearn.neighbors import KNeighborsRegressor
knnr = KNeighborsRegressor()
```

```python
knnr.fit(x_train_ss,y_train)
```

```python
y_pred = knnr.predict(x_test_ss)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
error_rate = []

for i in range(1,40) :
  knnr = KNeighborsRegressor(n_neighbors=i)
  knnr.fit(x_train, y_train)
  pred_i = knnr.predict(x_test)
  error_rate.append(mean_absolute_percentage_error(y_test, pred_i))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(range(1,40),error_rate,color='red', linestyle='--', marker='o',)
ax.set(xlabel = 'k', ylabel = 'test error', title='Elbow Chart for selection of k')
plt.show()
```
